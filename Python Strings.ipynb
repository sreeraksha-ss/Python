## 2.Accessing elements in a string

- Elements present/non-present can be checked using  Memership operators[in / not in]
- Elements can be accessed using indexing

```python
 print(a[4]) # indexing starts from zero hence the fourth element o is printed
```

Indexing can be done in python in two ways
- Forward indexing
- Backward indexing
    

Forward indexing allows us to index from left of the string and indexing starts from 0



Backward indexing allows us to index from right to left with a negative sign(in negative indexing ,the value starts from 1)

```python
# Example 1.0

a = 'python and Data Science' # it is saved contiguously in memory allocation
print(a[3])    # ---> forward indexing
print(a[-5])   # ---> Negative indexing
```

## 3.Inserting values in a string Data structure

As String is *Immutable* it is not possible to add  or update any value in the string(replacing is possible using methods)

```python
a = 'python and Data Science'
a[15]= 'k'
print(a)
```

Note that in the same memory allocation we can replace the string element using only methods

a.((tab-->intelligence on methods)replace(#shift +tab)---> to know about parameters

```python
a = 'python and Data Science'
print(a)
a = a.replace('n', 'S')
print(a)
```

## 4.Slicing in string data Structure

Slicing concept is used when a particular area of intrest is necessary and we can access only those elemnts in a string

slicing uses indexing concept to slice the strings into small sequence of charactors(we can not only slice single charectors but multi charactors also)

Note: Even space is considered as
Syntax for slicing is string[start:stop:step]

```python
a = 'python and Data Science'
# selecting the entire string
a[:]
```

```python
a = 'python and Data Science'
#selecting a[i] to a[j] index exluding jth index
a[1:5]
```

```python
a = 'python and Data Science'
#selecting first to seventh element
a[0:8]
```

```python
a = 'python and Data Science'
# Backward indexing
a[-10:-15]
```

```python
a = 'python and Data Science'#slicing in steps of 3
a[3:20:4]
```

##  5.Traversing through the Strings

Traversing means Visiting each element once through out the string

This can be done using indexing and for loop concepts

```python
a = 'python and Data Science'
for i in range(len(a)):
    print(a[i])
```

## 6.Search for an element in the string

```python
a = 'python and Data Science'
#To check if a single charactor exists,Use in operator ---> this returns boolean value
print ('n' in a)
```

```python
a = 'python and Data Science'

for i in range(len(a)):
    if a[i] == 'n':
        continue
print(a.count('n'))
```

## 7.Deletion in String

We know that deletion of an element in a string is not possible as the data structure is immutable

but is it possible to delete an entire string?


```python
a = 'python and Data Science'
del a
print(a)
# we can see a name error,that means it isnot possible to delete an element in string but we can delete entire string using del keywords
```

```python
a = 'python and Data Science'
for i in range(len(a)):
    del a[i]
```
