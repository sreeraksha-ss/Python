### 3.Accessing/Traversing the Dictionary
* Accessing value can be done using get method
  Syntax --> [dictionary.get('key' , 'value')
      * value is optional in the syntax
        dict['key']-->  outputs value if present

      * dict.get('key') --> outputs value if presesnt

      * dict['key not present']---> outputs key error

      * dict.get('key not present')---> outputs ** None **
* Traversing can be done with the index using key

```python
# Accessing Dictionary

# method1
# Syntax ---> dict['key']--outputs value if present
dict1={'one': 'si', 'dos': 'gracios', 'three': 'sing'}
print(dict1['dos'])
```

```python
# method2
# Syntax ---> dict.get(['key']) -- outputs value if presesnt
dict1={'one': 'si', 'dos': 'gracios', 'three': 'sing'}
print(dict1.get('dos'))
```

```python
# method3
# Syntax ---> dict['key not present']-- outputs key error
dict1={'one': 'si', 'dos': 'gracios', 'three': 'sing'}
print(dict1['tres'])
```

```python
# method4
# Syntax ---> dict.get('key not present')---> outputs ** None **
dict1={'one': 'si', 'dos': 'gracios', 'three': 'sing'}
print(dict1.get('tres'))
```

### 4.Update/Insert a Dictionary in python

### 6.Deletion methods in Dictionary

**Deletion of elements/Dictionary can be done using pop() ,delete() and popitem()**

```python
#using del keyword
#syntax--> del dict1['key'] -- Deletes key value pair from dictionary
dict1 = {'one': 'si', 'dos': 'gracios', 'three': 'sing', 1: 'ab', 2: 'bc', 3: 'cd'}
del dict1[2]
dict1
```

```python
#using del keyword  we can delete entire dictionary
#syntax--> del dict1 -- Deletes  dictionary
dict1 = {'one': 'si', 'dos': 'gracios', 'three': 'sing', 1: 'ab', 2: 'bc', 3: 'cd'}
del dict1
dict1
```

```python
#pop('key')--> removes key-value in the Dictionary
#syntax--> Dictionary.pop('key')
dict1= {'one': 'si', 'dos': 'gracios', 'three': 'sing', 1: 'ab', 2: 'bc', 3: 'cd'}
dict1.pop('one')
dict1
```

```python
#popitem()--> removes  last inserted key-value pair in the Dictionary
#syntax--> Dictionary.pop('key')
dict1= {'one': 'si', 'dos': 'gracios', 'three': 'sing', 1: 'ab', 2: 'bc', 3: 'cd'}
dict1.popitem()
dict1
```

 ### 6.Dictionary Methods

 1.clear()    - Deletes all key-value pair from dictionary
             [syntax: dict.clear()]

 2.copy()     - creates shallow copy of dictionary
             [syntax: dict.copy()]

 3.fromkeys() - creates new dictionary from given sequence of elements with values provided by the user
            [syntax: dict.fromkeys(sequence[1,2,3],0)]

 4.get()      - Returns the value for specified key , valueif key is in the dictionary
            [syntax: dict.get(key ,value)]

 5.items()    - View object that displays list of dictionaries,key-value, tuple pairs
            [syntax: dict.items()]

 6.keys()    - View object that displays list of all keys in the dictionary,
            [syntax: dict.keys()]
          
 7.values()    - View object that displays list of all values in the dictionary,
            [syntax: dict.values()]

 8.setdefault()    - Returns value of key if key is in dictionary if not it inserts key with value to the dictionary
            [syntax: dict.Setdeafult(key_default,value)]

 9.pop()    - Removes element and returns the element from the dictionary having given key
            [syntax: dict.pop('key')]

 10.popitem()    - removes an arbitrary element or last inserted element from dictionary
            [syntax: dict.popitem()]

 11.update()    - Updates the dictionary with elements from another dictionary or object from an iterable of key value pairs
            [syntax: dict.update(dict1)]

----List and Dictionary----
 * similarities
1. Both data structures are mutable
2. Both can be  iterated through
3. Both contain all data type elements

---
 * differences
1. Slicing is possible in list but is not possible in dictionary
2. Dictionary is stored in key value pairs  but list are stored as non dependable elements
3.Accessing is done via keys in Dictionary whereas in list it is done through indexes
4. Dictionary does not allow duplicate keys where as list allows duplicate values
