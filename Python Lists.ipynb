### 3.Accessing /Traversing the list

Accesing the element can be done either using forward indexing or reverse indexing

```python
# Accessing

nts= [1,2,'ab',12.5]
print(nts[0])  # forward indexing
print(nts[-2]) # backward indexing
```

```python
#Traversing -->visiting each element atleast once in the list (traversing can be done using for loop)

#method 1 using membership operator in
nts= [1,2,'ab',12.5]
for i in nts:
  print(i)
```

```python
#method 2 using range and len functions

for i in range(len(nts)):
  nts[i] += nts[i] ,+, str[1]
  print(nts[i])
```

### 4.Update/Insert a list in python

```python
# Updating elements --> without changing the memoruy location we update a value to a given location

list1=[12,35,'hgf',13.65, 854]
list1[4]='bmw'
list1
```

```python
# Updating the elements in place can also be done using slicing operator
# we will learn about slice in next section

list1=[12,35,'hgf',13.65, 854]
list1[0:2]= ['z', 15]
print(list1)
```

**Inserting elements in python list (insert,append,extend)**

Using insert method we can insert value anywhere inthe list using indexing

```python
#1.Inserting  single Element at the beginning/any given index value in the of the list
# syntax =[list1.insert(index, value)]

list1=[12,35,'hgf',13.65, 854]
list1.insert(1,'ab')
list1
```

```python
# 2.Append --> used to add single at the end of the list
# Syntax -->(list1.append(value)) # if Str-->'Value'
list1=[12,35,'hgf',13.65, 854]
list1.append('ab1')
list1
```

```python
# 2.Append --> does not takes multiple values to append
list1=[12,35,'hgf',13.65, 854]
list1.append('ab1','bab1')
list1
```

```python
#extend method is used to add list to the list,or more values
#syntax-->(mylist.extend([5,6,7,8]))

list1=[12,35,'hgf',13.65, 854]
list1.extend(['ab1','bab1'])
list1
```

```python
# Extend method can add nested list inside a list too
list1=[12,35,'hgf',13.65, 854]
list1.extend([['ccv','ccb1']])
list1
```

### 6.Deletion methods in list

**Deletion of elements/list can be done using pop() ,delete() and remove()**

```python
#pop(i)--> removes the ith index element in the list
#syntax--> list.pop(index)
list1=[12,35,'hgf',13.65, 854]
list1.pop(2)
list1
```

```python
#pop()--> removes the last element in the list
#syntax--> list.pop()
list1=[12,35,'hgf',13.65, 854]
list1.pop()
list1
```

```python
# delete method works on index and slicing
#syntax--> del list[index]
list1=[12,35,'hgf',13.65, 854]
del list1[3]
list1
```

```python
# delete method works on index and slicing
#syntax--> del list[0:n]
list1=[12,35,'hgf',13.65, 854]
del list1[0:3]
list1
```

```python
#delete method works on entire list
#syntax--> del list
list1=[12,35,'hgf',13.65, 854]
del list1
list1
```

```python
#Remove method (works on the element to be remooved if the index value is unknown)
#syntax--> list.remove('element')

list1=[12,35,'hgf',13.65, 854]
list1.remove(35)
list1
```

```python
#Remove method (works on the element to be remooved if the index value is unknown)
#syntax--> list.remove('element') #it does not take more than one argument

list1=[12,35,'hgf',13.65, 854]
list1.remove([12,35])
list1
```

```python
##Searching an elemnet in the list
#1)Using in operator

list1=[12,35,'hgf',13.65, 854]
target= 13.65
if target in list1:
  print(f"{target} is in the list")
else:
  print(f"{target} is not in the list")
```

```python
#2)Using linear search
list1=[12,35,'hgf',13.65, 854]
def linearsearch(list1 ,target):
  for i , value in enumerate(list1):   # i takes index,value takes element
      if  value == target:
        return i
  return -1
print(linearsearch(list1,854))
```
