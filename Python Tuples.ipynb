### 3.Accessing /Traversing the Tuple

Accesing the element can be done either using forward indexing or reverse indexing

```python
# Accessing

nts= (1,2,'ab',12.5)
print(nts[0])  # forward indexing
print(nts[-2]) # backward indexing
print(type(nts))
```

```python
#Traversing -->visiting each element atleast once in the Tuple (traversing can be done using for loop)

#using membership operator in
nts= (1,2,'ab',12.5)
for i in nts:
  print(i)
```

### 5.Slicing  Concept in Tuple
slicing is possible only to access or traverse but not to update or delete
    syntax-->(Tuple1(start:stop:step))

```python
Tuple1=(12,35,'hgf',13.65, 854)
Tuple1
```

```python
# first two elements
print(Tuple1[0:])
```

```python
#second element to len of Tuple
print(Tuple1[1:])
```

```python
# All elements
print(Tuple1[:])
```

```python
#In steps of two
print(Tuple1[::2])
```

```python
#reversing the Tuple elements
print(Tuple1[::-1])
```

### 6.Deletion methods in Tuple

```python
#delete method works on entire Tuple
#syntax--> del Tuple
Tuple1=(12,35,'hgf',13.65, 854)
del Tuple1
Tuple1
```

```python
##Searching an elemnet in the Tuple
#1)Using in operator

Tuple1=(12,35,'hgf',13.65, 854)
target= 13.65
if target in Tuple1:
  print(f"{target} is in the Tuple")
else:
  print(f"{target} is not in the Tuple")
```

```python
#2)Using linear search
Tuple1=(12,35,'hgf',13.65, 854)
def linearsearch(Tuple1 ,target):
  for i in enumerate(Tuple1):   # i takes index,value takes element
      if  i == target:
        return i
  return -1
print(linearsearch(Tuple1,854))
```

```python
#len()--> returms total number of elements in the Tuple
#max()--> Returms max value in the Tuple
#min()--> returns minimum value in the Tuple
#sum()--> Returns sum value in a Tuple
#average of Tuple is sum(Tuple)/len(Tuple)
#index()--> returns index value of the given tuple
#count()--> returns number of occurances of a value in a tuple
Tuple1=(12,35,13.65, 854)
print (f'Length of Tuple is {len(Tuple1)}')
print (f'max of Tuple is {max(Tuple1)}')
print (f'min of Tuple is {min(Tuple1)}')
print (f'sum of Tuple is {sum(Tuple1)}')
print (f'sum of Tuple is {sum(Tuple1)/len(Tuple1)}')
```

# differences

* Tuples are immutable|Lists are mutable
* Tuples can be renamed but cannot update, replace, delete, or add  elements wheres| lists can do all the operations
* Slicing in tuple is possible only to access or traverse | lists can change elements in lists using slicing concept
* Tuples are usually opted for heterogeneous datatypes | Lists are usually opted for homogenious datatypes.
