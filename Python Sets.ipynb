### 4.Update/Insert a Set in python
* Values inside a set are immutable data types hence they are not subscriptible
* we cannot update an element in the set
* but with different methods avaliable we can add elements to the set for example we can insert multiple values to the set using update() method in the set

```python
# Updating elements --> without changing the memory location we update a value to a given location

Set1=({12,35,'hgf',13.65, 854})
Set1[5]= ({6})
Set1
```

**Inserting elements in python Set (add(), update()**
 * add() --> to add single element in the set
 * update() --> to add multiple elements in the set

```python
#1.Inserting/adding  single Element
# syntax =Set1.add({value)})

Set1=({12,35,'hgf',13.65, 854})
Set1.add('ab')
Set1
```

```python
# update method
Set1=({12,35,'hgf',13.65, 854})
Set1.update({'ccv','ccb1'})
Set1
```

### 6.Deletion methods in Set

**Deletion of elements/Set can be done using discard() ,delete() clear()**

* Discard()-> removes an element from the set
* pop()--> removes arbitrary element in the Set
* clear()  -> removes all elements from the set
* del      -> deletes the entire set


```python
#Discard()-> removes an element from the set
#syntax--> Set.discard(element)
Set1=({12,35,'hgf',13.65, 854})
Set1.discard(35)
Set1
```

```python
#pop()--> removes arbitrary element in the Set
#syntax--> Set.pop()
Set1=({12,'hgf',13.65, 854})
Set1.pop()
Set1
```

```python
# delete method works on entire set
#syntax--> del Set({index})
Set1=({12,35,'hgf',13.65, 854})
del Set1
Set1
```

```python
#clear method (be removes the entire set elements and throws an empty set )
#syntax--> Set.clear()

Set1=({12,35,'hgf',13.65, 854})
Set1.clear()
Set1
```

```python
#Remove method (throws an error if element dosentexist in the set hence discard is used)
#syntax--> Set.remove('element') #it does not take more than one argument

Set1=({12,35,'hgf',13.65, 854})
Set1.remove(852)
Set1
```

```python
##Searching an elemnet in the Set
#1)Using in operator

Set1=({12,35,'hgf',13.65, 854})
target= 13.65
if target in Set1:
  print(f"{target} is in the Set")
else:
  print(f"{target} is not in the Set")
```

```python
#2)Using linear search
Set1=({12,35,'hgf',13.65, 854})
def linearsearch(Set1 ,target):
  for i , value in enumerate(Set1):   # i takes index,value takes element
      if  value == target:
        return i
  return -1
print(linearsearch(Set1,854))
```

### Set Operations/functions

* Union() appends two sets


```python
# union()-> appens two sets

a=({1,2,3})
b=({4,7,9})
a.union(b)
```

```python
# Intersection --> view object that shows common elements in 2 sets
a=({1,2,3,4, 6 ,9})
b=({4,7,9})
a.intersection(b)
```

```python
# Set difference
# a-b - elements that are present in a but not in b |(also written as a.difference(b))
a=({1,2,3,4, 6 ,9})
b=({4,7,9})
print(a-b)
print(b.difference(a))
```

```python
# Symmetric difference [Union-Intersection]
a=({1,2,3,4, 6 ,9})
b=({4,7,9})
print(a ^ b)
print(a.symmetric_difference(b))
```
