**List and Dictionary comprehension Makes the code easily readable and cuts the lines of code **

List Comprehension takes input from sequence data types
i.e, list ,range,tuple


Note that while writing syntax there are no commas in between like regular coding


---
Syntax: newlist=[(print /operation ) for  item in data] # no condtion

```python
# normal code
prelist=[1,2,3,4]
newlist=[]
for i in prelist:
  mul = i*2
  newlist.append(mul)
print(newlist)
```

```python
# Using the same operation these few lines of code can be put in one line

prelist=[1,2,3,4]
b= [i*2 for i in prelist]
print(b)
```

```python
a = (11,23,45,67)
b = [i // 7  for i in a if i>40]
b
```

### Dictionary Comprehension

* Dictionary Comprehension takes input from sequence data types  list and dictionary

* Note that while writing syntax there are no commas in between like regular coding
---
Syntax: newdict={newkey : new value for item in list/dict.items()} # no condtion

```python
prelist=[1,2,3,4]
b= {i: i*2 for i in prelist}
print(b)
```

```python
predict={1: 100,2: 200,3: 300,4:400}
b= {i*10 :j for i,j in predict.items()}
print(b)
```

#### Syntax of Dictionary comprehension **using if and for loop**

* Dictionary = {j:  i**2 for i,j  in data if i % 2 == 0}
* use if to the right of for loop

```python
#normal code
a={'data': 100 , 'science': 150 , ' Ai': 200}
c= {}
for i , j in a.items():
  if j>100:
    c[i] = j / 2
print(c)
```

```python
# using Dictionary comprehension
a={'data': 100 , 'science': 150 , ' Ai': 200}
c = { i: j/2 for i, j in a.items() if j > 100}
print(c)
```

### Syntax of Dictonary Comprehension using **if , else , and for loop**
* Use if condition, then else condtition , before using for loop
* Syntax: dict= {-print/operation---if result---if condition---else---else condition---for condition}

```python
a={'data': 100 , 'science': 150 , ' Ai': 200}
c = { i: j/2  if j > 100 else i : j + 3.33 for i, j in a.items()}
print(c)
```
